```python
from collections import deque

class Graph:
    def __init__(self, graph, heuristic):
        self.graph = graph  # Store the graph as an adjacency list
        self.heuristic = heuristic  # Heuristic for Best First Search

    def best_first_search(self, start_node, goal_node):
        open_list = [start_node]  # Nodes to be evaluated
        closed_list = []  # Nodes that have been evaluated
        came_from = {}  # To store the best path from the start to the goal

        while open_list:
            # Select the node with the lowest heuristic value (best estimate)
            current_node = min(open_list, key=lambda node: self.heuristic[node])
            open_list.remove(current_node)  # Remove the node from open list
            closed_list.append(current_node)  # Add to closed list

            print(f"Iteration {len(closed_list)}: Open {open_list}, Closed {closed_list}")

            # If the goal node is reached, reconstruct the path
            if current_node == goal_node:
                path = []
                while current_node in came_from:
                    path.append(current_node)
                    current_node = came_from[current_node]
                path.append(start_node)
                path.reverse()
                print(f"Solution Path: {' --> '.join(path)}")
                return path

            # Expand the neighbors of the current node
            for neighbor in self.graph[current_node]:
                if neighbor not in closed_list:
                    open_list.append(neighbor)
                    came_from[neighbor] = current_node  # Record where each node came from

        print("No solution found.")
        return None

# Define the graph as an adjacency list (example graph with node connections)
graph_data = {
    'S': ['B', 'A'],
    'A': ['C', 'D'],
    'B': ['S', 'F', 'G'],
    'C': ['A', 'E'],
    'D': ['A'],
    'E': ['C'],
    'F': ['B'],
    'G': ['B']
}

# Define the heuristic for each node (used for Best First Search)
# Heuristic values represent the cost or estimated distance to the goal
heuristic_values = {
    'S': 1,
    'A': 3,
    'B': 5,
    'C': 2,
    'D': 7,
    'E': 6,
    'F': 1,
    'G': 0
}

# Create a Graph object with the graph data and heuristic values
graph = Graph(graph_data, heuristic_values)

# Define the start and goal nodes
start_node = 'S'
goal_node = 'G'

# Perform Best First Search
print(f"Best First Search starting from node {start_node} to reach node {goal_node}:")
graph.best_first_search(start_node, goal_node)



```
