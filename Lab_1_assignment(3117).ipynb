```python
import random

class GuessingGame:
    def __init__(self, lower_bound=1, upper_bound=250):
        self.lower_bound = lower_bound
        self.upper_bound = upper_bound
        self.secret_number = random.randint(self.lower_bound, self.upper_bound)
        self.guesses_taken = 0

    def start_game(self):
        print(f"I'm thinking of a number between {self.lower_bound} and {self.upper_bound}.")
        self.play_game()

    def play_game(self):
        while True:
            try:
                guess = int(input("Enter your guess: "))
                self.guesses_taken += 1  # Count each guess

                # Check if the guess is correct
                if guess < self.secret_number:
                    print("Too low! Try again.")
                elif guess > self.secret_number:
                    print("Too high! Try again.")
                else:
                    print(f"Congratulations! You guessed the number in {self.guesses_taken} guesses.")
                    break  # Exit the loop when the correct guess is made
            except ValueError:
                print("Please enter a valid number.")

# Start the game
if __name__ == "__main__":
    game = GuessingGame()
    game.start_game()
```
