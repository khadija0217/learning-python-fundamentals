```python
# 1. Importing deque
from collections import deque

# 2. Defining the Graph class
class Graph:
    def __init__(self, graph):
        self.graph = graph  # Store the graph as an adjacency list

    # 3. Defining the depth-limited search (DLS) method
    def dls(self, start_node, depth_limit):
        visited = set()  # To keep track of visited nodes
        stack = [(start_node, 0)]  # Stack contains tuples of (node, current_depth)

        # 4. Processing nodes in the stack
        while stack:
            node, depth = stack.pop()  # Get the current node and its depth

            # 5. Stop processing if the depth limit is reached
            if depth > depth_limit:
                continue

            # 6. If the node is not visited, process it
            if node not in visited:
                print(node, end=" ")  # Print the node
                visited.add(node)  # Mark it as visited

                # 7. Push unvisited neighbors onto the stack with incremented depth
                for neighbor in reversed(self.graph[node]):
                    if neighbor not in visited:
                        stack.append((neighbor, depth + 1))

# 8. Creating a Graph object and calling dls
# Define the graph as an adjacency list
graph_data = {
    'k': ['H','A', 'D'],
    'H': ['I'],
    'A': ['J', 'A'],
    'D': ['H'],
    'I': ['T'],
    'J': [],
    'A': [],
    'T':[],
    'H': ['A','H'],
    'A':['I','R'],
    'I': [],
    'R': []
}

# Create a Graph object and call the dls method
graph = Graph(graph_data)

# Specify the start node and depth limit
start_node = 'K'
depth_limit = 3

print(f"Depth-Limited Search Traversal starting from node {start_node} with depth limit {depth_limit}:")
graph.dls(start_node, depth_limit)
```
